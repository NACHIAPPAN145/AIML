# Random Forest

```python
# importing required libraries
# importing Scikit-learn library and datasets package
from sklearn import datasets

# Loading the iris plants dataset (classification)
iris = datasets.load_iris()
```

```python
print(iris.target_names) # name of the target , 3 label names target has
```

```python
print(iris.feature_names) # name of the columns that are used to find target ,4 features
```

```python
# dividing the datasets into two parts i.e. training datasets and test datasets
X, y = datasets.load_iris( return_X_y = True)

# Splitting arrays or matrices into random train and test subsets
from sklearn.model_selection import train_test_split
# i.e. 70 % training dataset and 30 % test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)
```

```python
# importing random forest classifier from assemble module
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
# creating dataframe of IRIS dataset
data = pd.DataFrame({'sepallength': iris.data[:, 0], 'sepalwidth': iris.data[:, 1],
                     'petallength': iris.data[:, 2], 'petalwidth': iris.data[:, 3],
                     'species': iris.target})
```

```python
print(data.head())
```

```python
# creating a RF classifier
clf = RandomForestClassifier(n_estimators = 100) # no of decision tree that has to build based on this data for this random forest

# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
clf.fit(X_train, y_train)

# performing predictions on the test dataset
y_pred = clf.predict(X_test)

# metrics are used to find accuracy or error
from sklearn import metrics  # quantifing the accuracy of model
print()

# using metrics module for accuracy calculation
print("ACCURACY OF THE MODEL:", metrics.accuracy_score(y_test, y_pred))  # the accuracy of model differs since random forest
```

```python
# predicting which type of flower it is.  # this also varies since the random forest has different voting based on ind decision tree generated
clf.predict([[3, 3, 2, 2]])  # 0 , 1, 2 , 3 array no - array prediction is 2 that means virginica species it belongs to
```

```python
# using the feature importance variable
import pandas as pd
feature_imp = pd.Series(clf.feature_importances_, index = iris.feature_names).sort_values(ascending = False)
feature_imp  # which feature is more important for finding target here petal width
```
